Wczytujemy biblioteki na potrzeby naszych zajęć.

```python
import pandas as pd
```

Wczytujemy dane z pliku xlsx z wykorzystaniem funkcji `read_excel`.

```python
dane=pd.read_excel("rent-poznan.xlsx")
dane.head()
```

```python
dane.tail(n = 10)
```

Ile wierszy i kolumn ma nasz zbiór danych?

```python
print(dane.shape[0])
print(dane.shape[1])
dane.shape[0], dane.shape[1]
```

Podsumowanie zbioru danych możliwe jest z wykorzystaniem funkcji `describe()`

```python
dane.describe()
```

Wybór jednej, określonej kolumny.

```python
dane.price
dane["price"]
dane[ ["price"] ]
dane[ ["price", "flat_area"] ]
dane[ ["price", "flat_area"] ].describe()
dane.price.describe()
```

Filtrowanie wierszy

```python
dane[ dane["price"] < 500 ][ ["id", "price", "ad_title"] ]
```

```python
dane_bledy = dane[ (dane["price"] < 1000) & (dane["flat_area"] < 15)][ ["id", "price", "ad_title"] ]
```

```python
dane_bledy.head(n=10)
```

Wybieramy ogłoszenia, których cena była niższa od 1000 i powierzchnia większa od 50 lub mniejsza od 10

```python
dane[ (dane["price"] < 1000) & ( (dane["flat_area"] >= 50) | (dane["flat_area"] <= 10) )][
    ["id", "price", "flat_area", "ad_title"]
]
```

```python
dane[ (dane["price"] < 1000) & (dane["flat_area"].between(40,60))][
    ["id", "price", "flat_area", "ad_title"]
]
```
